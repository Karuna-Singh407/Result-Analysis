# Drop unnamed value

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
df.head()
```

```python
df["TransportMeans"] = df["TransportMeans"].fillna("other_means")
```

```python
df.head()
```

# Gender Distribution

```python
gender_counts=df["Gender"].value_counts()
```

```python
gender_counts
```

```python
c=["orange","b"]
plt.figure(figsize=(6,5))
plt.bar(gender_counts.index,gender_counts.values,color=c,label=gender_counts.values)
plt.title("Gender Distribution",fontsize=15)
plt.xlabel("Gender",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.ylim(0,20000)
plt.legend()
plt.show()
          
```

```python
#From the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's education and student's score")
plt.show()
```

```python
# From the above chart we have concluded that the education of the parents have a good impact on the students score 
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Marital status and student's score")
plt.show()
```

```python
# From the above chart we have concluded that the marital status of the parents have nno/neglegible impact on the students score
```
